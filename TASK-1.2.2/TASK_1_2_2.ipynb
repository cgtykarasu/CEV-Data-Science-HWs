**First 100 prime numbers: use for instead of while**

```python
####################################
#       DATA SCIENCE BOOTCAMP      #
# File name : TASK-1.2.2.ipynb     #
# Author    : Çağatay KARASU       #
####################################

counter = 1
num = 3
my_prime_list = [2]

def is_prime (num):
  i = 3
  if (num<2):
    return False
  elif (num==2):
    return True
  elif(num%2 == 0):
    return False
  while(i < (num//2)):
    if (num % i == 0):
      return False
    i += 2
  return True

for i in range(3,1000,2):
  if(is_prime(i)):
    my_prime_list.append(i)
    counter += 1
    if(counter == 100):
      break

for i in range(100):
  print('{}.prime number : {}'.format(i+1, my_prime_list[i]))
```
