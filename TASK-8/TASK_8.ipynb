```python
####################################
#      DATA SCIENCE BOOTCAMP       #
#    File name : TASK-8.ipynb      #
# Author    : Çağatay KARASU       #
####################################

import time
import random

class Kumanda():
  def __init__(self, tv_durum="Kapalı", tv_ses=0, kanal_listesi=["FOX"], kanal="FOX", internet="Bağlı Değil!", Netflix="Kapalı", Netflix_odeme="Kapalı", ses_stereo_mode="Kapalı"):
    self.tv_durum = tv_durum
    self.tv_ses = tv_ses
    self.kanal_listesi = kanal_listesi
    self.kanal = kanal
    self.internet = internet
    self.Netflix = Netflix
    self.Netflix_odeme = Netflix_odeme
    self.ses_stereo_mode = ses_stereo_mode

  def tv_ac(self):
    if (self.tv_durum  == "Kapalı"):
        print("Tv açılıyor...")
        self.tv_durum="Açık"
    else:
        print("Tv zaten açık!")

  def tv_kapat(self):
    if (self.tv_durum  == "Kapalı"):
        print("Tv zaten kapalı!")
        
    else:
        print("Tv kapanıyor...")
        self.tv_durum="Kapalı"
    

  def ses_ayari(self):
    while True:
        cevap = input("Sesi artır: >\nSesi azalt: <\nÇıkış: q")

        if (cevap=='>'):
            if self.tv_ses>=100:
                print("Ses maksimumda!")

            self.tv_ses += 1
            print("Tv Sesi: {}".format(self.tv_ses))
            
        elif (cevap=="<"):
            if (self.tv_ses<=0):
                print("Ses minimumda!")

            self.tv_ses-=1
            print("tv_ses: {}",format(self.tv_ses))

        elif (cevap=="q"):
            break
        else:
            print("Hatalı giriş!")

  def kanal_ekle(self,kanal):
    print("Kanal ekleniyor...")
    time.sleep(1)
    self.kanal_listesi.append(kanal)

    print("Kanal listesi: {}".format(self.kanal_listesi))

  def kanal_sec(self):
    rnd = random.randint(0,len(self.kanal_listesi)-1)
    self.kanal = self.kanal_listesi[rnd]
    print(self.kanal)
    
  def __len__(self):
    return len(self.kanal_listesi)

  def Connect_Netflix(self):
    while True:
        if self.Netflix == "Açık":
            print("Netflix zaten açık!")
            break
        elif self.Netflix == "Kapalı":
            cevap = input("Netflix'e bağlanmak istiyor musunuz? (E/H)")
            if cevap == "E" and self.internet == "Bağlı Değil!":
                print("İnternet Bağlı Değil!")
                cevap2 = input("İnternete Bağlanmak İstiyor musunuz? (E/H)")
                if cevap2 == "E":
                    print("İnternete Bağlanıyor...")
                    time.sleep(3)
                    self.internet = "Bağlı!"
                    print("İnternete Bağlandınız!")
                    print("Netflix'e Bağlanılıyor...")
                    time.sleep(1)
                    self.Netflix = "Açık"
                    print("Netflix Açıldı!")
                    break
                elif cevap2 == "H":
                    print("İnternet olmadan Netflix hizmetine bağlanamazsınız!")
                    time.sleep(1)
                    break
                else:
                    print("Lütfen E veya H harflerinden birini giriniz.")
                    continue
            elif cevap == "E" and self.internet == "Bağlı!":
                print("İnternete bağlanıldı, Netflix Açılıyor...")
                self.Netflix = "Açık"
                time.sleep(1)
                print("Netflix Açıldı!")

            elif cevap == "H":
                print("Çıkış yapılıyor...")
                time.sleep(1)
                break

            else:
                print("Lütfen E veya H harflerinden birini giriniz.")
                continue

  def Netflix_odeme(self):
    while True:
      cevap = input("Netflix abonelik ödemesi yapmak istiyor musunuz? (E/H)")
      if cevap == "E":
        if self.Netflix_odeme == "Kapalı":
          print("Netflix abonelik ödemesi açılıyor...")
          self.Netflix_odeme = "Açık"
          break

        else:
          print("Netflix abonelik ödemesi zaten yapıldı.")
          break

      elif cevap == "H":
        print("Netflix abonelik ödemesini reddettiniz, sistem kapatılıyor...")
        break

      else:
        print("Lütfen E veya H harflerinden birini giriniz.")
        continue

  def ses_stereo_mode(self):
    while True:
      cevap = input("Stereo ses modunu açmak istiyor musunuz? (E/H)")
      if cevap == "E":
        if self.ses_stereo_mode == "Kapalı":
          print("Stereo ses modu açılıyor...")
          self.ses_stereo_mode = "Açık"
          break

        else:
          print("Stereo ses modu zaten açık!")
          break

      elif cevap == "H":
        print("Stereo ses modunu açmayı reddettiniz, sistem kapatılıyor...")
        break

      else:
        print("Lütfen E veya H harflerinden birini giriniz.")
        continue

print("""
        1. TV aç
        2. TV kapat
        3. Ses Ayarları
        4. Kanal Ekle
        5. Açık Kanalı Öğren
        6. Kanal Sayısı
        7. TV Bilgileri
        8. Netflix'e Bağlan
        9. Netflix ödemesi yap
        10.Stereo ses modunu aç
        Çıkmak için q'ya bas.
""")

kumanda = Kumanda()

while True:
  islem= input("Lütfen yapmak istediğiniz işlemi seçiniz:")

  if islem == "q":
    print("Sonlandırılıyor")
    break 
  elif islem == "1":
    kumanda.tv_ac()
    
  elif islem == "2":
    kumanda.tv_kapat()

  elif islem == "3":
    kumanda.ses_ayari()
    
  elif islem == "4":
    kanal_isimleri=input("Kanal isimlerini lütfen , ile ayırarak giriniz.")
    x = kanal_isimleri.split(",")

    for i in x:
        kumanda.kanal_ekle(i)
    
  elif islem == "5":
    kumanda.kanal_sec()

  elif islem == "6":
    print("Kanal sayısı: ",len(kumanda))
    
  elif islem == "7":
    print(kumanda)

  elif islem == "8":
    kumanda.Connect_Netflix()

  elif islem == "9":
    kumanda.Netflix_odeme()

  elif islem == "10":
    kumanda.ses_stereo_mode()

  else:
    print("hatalı Tuşlama")
```

```python
class PCLab():
  def __init__(self, sinif_adi = "CEV&BEBKA - Veri Bilimi", bilgisayar_sayisi = 10, doluluk_durumu = "Boş", ogrenci_sayisi = 10, asistan = ["Çağatay"]):
    self.sinif_adi = sinif_adi
    self.bilgisayar_sayisi = bilgisayar_sayisi
    self.ogrenci_sayisi = ogrenci_sayisi
    self.doluluk_durumu = doluluk_durumu
    self.asistan = asistan

  def __len__(self):
    return len(self.asistan)

  def bilgisayar_ekle(self):
    while True:
      cevap = input("Labdaki Mevcut bilgisayar sayısı : {},\
      \nbilgisayar sayısını 1 arttırmak için '+'\
      \nBilgisayar sayısını 1 azaltmak için '-'\
      \nÇıkış için 'q':".format(self.bilgisayar_sayisi))

      if cevap == "+":
        if self.bilgisayar_sayisi != 25:
          self.bilgisayar_sayisi += 1
          print("Bilgisayar Sayısı: {}".format(self.bilgisayar_sayisi))

        elif self.bilgisayar_sayisi == 25:
          print("Maksimum bilgisayar sayısına ulaşıldı!")

      elif cevap == "-":
        if self.bilgisayar_sayisi != 0:
          self.bilgisayar_sayisi -= 1
          print("Bilgisayar Sayısı: {}".format(self.bilgisayar_sayisi))

        elif self.bilgisayar_sayisi == 0:
          print("Bilgisayar sayısı sıfır!")

      else:
        print("Mevcut Bilgisayar Sayısı: {}".format(self.bilgisayar_sayisi))
        break

  def lab_doluluk_durumu(self):

    if (self.ogrenci_sayisi >= self.bilgisayar_sayisi):
      print("Tüm bilgisayar kullanılmakta!")
      self.doluluk_durumu="Dolu"

    else:
      print("Boş bilgisayar bulunmakta.")
      self.doluluk_durumu="Boş"

  def ogrenci_ekle(self):
    while True:
      cevap = input("Labdaki Mevcut Öğrenci sayısı: {},\
      \nöğrenci sayısını 1 arttırmak için '+'\
      \nÖğrenci sayısını azaltmak için '-'\
      \nÇıkış için 'q'\n".format(self.ogrenci_sayisi))

      if cevap == "+":
        if self.ogrenci_sayisi != 50:
          self.ogrenci_sayisi += 1
          print("Öğrenci Sayısı:", self.ogrenci_sayisi)
        else:
          print("Daha fazla öğrenci eklenememekte.")

      elif cevap == "-":
        if self.ogrenci_sayisi != 0:
          self.ogrenci_sayisi -= 1
          print("Öğrenci Sayısı:", self.ogrenci_sayisi)
        else:
          print("Daha fazla çıkarılamaz.")
      else:
        print("Labdaki Mevcut Öğrenci sayısı: {}".format(self.ogrenci_sayisi))
        break

  def asistan_ekle(self):
    asistan_adi = input("Eklemek istediğiniz asistanın adını giriniz...")
    print("Asistan Sınıfa görevlendiriliyor...")
    self.asistan.append(asistan_adi)
    print("Asistan listesi: {}".format(self.asistan))

  def __str__(self):
    return "Sınıf Kodu: {}\
    \nBilgisayar Sayısı: {}\
    \nÖğrenci Sayısı: {}\
    \nDoluluk Durumu: {}\
    \nAsistanlar: {}"\
    .format(self.sinif_adi, self.bilgisayar_sayisi, self.ogrenci_sayisi, self.doluluk_durumu,self.asistan)

lab = PCLab()

print("""
    1. Bilgisayar Ekle
    2. Doluluk Durumu
    3. Öğrenci Ekle
    4. Asistan Sayısı
    5. Tüm Bilgiler
    6. Asistan Ekle
    7. Çıkış
""")

while True:
    cevap = input("Seçim Yapınız: ")
    if cevap == "1":
        lab.bilgisayar_ekle()
    elif cevap == "2":
        lab.lab_doluluk_durumu()
    elif cevap == "3":
        lab.ogrenci_ekle()
    elif cevap=="4":
        print("Asistan sayısı: ",len(lab.asistan))
    elif cevap=="5":
        print(lab)
    elif cevap=="6":
        lab.asistan_ekle()

    else:
        print("Çıkış Yapılıyor...")
        break
```

```python
class Animals():
    def __init__(self, name, genre, height, weight):
        self.name = name
        self.genre = genre
        self.height = height
        self.weight = weight

    def show_info(self):
        print("Name : {}\nGenre: {}\nHeight: {}\n Weight: {}".format(self.name,self.genre, self.height, self.weight))
```

```python
class Dinosaur(Animals):
    def __init__(self, name, genre, height, weight, color):
        super().__init__(name, genre, height, weight)
        self.color = color

    def show_info(self):
        print("Name : {}\nGenre: {}\nHeight: {}\nWeight: {}\nColor: {}".format(self.name,self.genre, self.height, self.weight, self.color))

D01 = Dinosaur("Dortonyanus","Raptor",3,100,"Green")
D01.show_info()
```

```python
class Tuna(Animals):
    def __init__(self, name, genre, height, weight, age):
        super().__init__(name, genre, height, weight)
        self.age = age

    def show_info(self):
        print("Genre: {}\nHeight: {}\nWeight: {}\nAge: {}".format(self.genre, self.height, self.weight, self.age))

O01 = Tuna("", "Katsuwonus pelamis", 4, 50, 25)
O01.show_info()
```

```python
class Eagle(Animals):
    def __init__(self, name, genre, height, weight, age, wing_length):
        super().__init__(name, genre, height, weight)
        self.age = age
        self.wing_length = wing_length

    def change_genre(self):
      print("Type the genre please: ")
      new_genre = input()
      self.genre = new_genre
    
    def show_info(self):
      print("Genre: {}\nHeight: {}\nWeight: {}\nAge: {}\nWing Length: {}".format(self.genre, self.height, self.weight, self.age, self.wing_length))


E01 = Eagle("", "Fish-eagles", 3, 10, 15, 1)
E01.show_info()
E01.change_genre()
E01.show_info()
```
