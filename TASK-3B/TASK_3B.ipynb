```python
####################################
#      DATA SCIENCE BOOTCAMP       #
# File name : TASK-3B.ipynb        #
# Author    : Çağatay KARASU       #
####################################

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data_all = pd.read_csv("MRI_and_CDRinfo_Values_X_train.csv")  #read this file and synchronize it "data_all"
data_all
```

```python
data_labels = pd.read_csv("CDR_Values_y_train.csv")           #read this file and synchronize it "data_labels"
data_labels["HIP_VOL"] = data_all["HIPPOVOL"]                 #take "HIPPOVOL" label from data_all and add it to data_labels
data_labels
```

```python
cdrblog_0, cdrblog_1, cdrblog_2, cdrblog_3 =  (data_labels[data_labels["CDRGLOB"]==0].iloc[:,2]), \
                                              (data_labels[data_labels["CDRGLOB"]==1].iloc[:,2]), \
                                              (data_labels[data_labels["CDRGLOB"]==2].iloc[:,2]), \
                                              (data_labels[data_labels["CDRGLOB"]==3].iloc[:,2])
```

```python
cdrglob_0_mean, cdrglob_1_mean, cdrglob_2_mean, cdrglob_3_mean =  cdrblog_0.mean(), \
                                                                  cdrblog_1.mean(), \
                                                                  cdrblog_2.mean(), \
                                                                  cdrblog_3.mean()

print('Hippocampus volume\'s mean for CDRBLOG 0 is {}'.format(cdrglob_0_mean))
print('Hippocampus volume\'s mean for CDRBLOG 1 is {}'.format(cdrglob_1_mean))
print('Hippocampus volume\'s mean for CDRBLOG 2 is {}'.format(cdrglob_2_mean))
print('Hippocampus volume\'s mean for CDRBLOG 3 is {}'.format(cdrglob_3_mean))
```

```python
data_labels.groupby('CDRGLOB')[['HIP_VOL']].mean()
```

```python
cdrglob_0_std, cdrglob_1_std, cdrglob_2_std, cdrglob_3_std =  cdrblog_0.std(), \
                                                              cdrblog_1.std(), \
                                                              cdrblog_2.std(), \
                                                              cdrblog_3.std()

print('Hippocampus volume\'s standard deviation for CDRBLOG 0 is {}'.format(cdrglob_0_std))
print('Hippocampus volume\'s standard deviation for CDRBLOG 1 is {}'.format(cdrglob_1_std))
print('Hippocampus volume\'s standard deviation for CDRBLOG 2 is {}'.format(cdrglob_2_std))
print('Hippocampus volume\'s standard deviation for CDRBLOG 3 is {}'.format(cdrglob_3_std))
```

```python
data_labels.groupby('CDRGLOB')[['HIP_VOL']].std()
```

```python
data_labels.groupby('CDRGLOB').aggregate({'HIP_VOL': ['mean','std']})
```

```python
pd.pivot_table(data_labels, index=['CDRGLOB'], values='HIP_VOL', aggfunc=['mean', 'std']).plot(kind="bar")
plt.title("Hippocampus volume\'s mean & standard deviation\nwith respect to the level of Alzheimer's Disease")
plt.xlabel("level of Alzheimer's Disease")
```
