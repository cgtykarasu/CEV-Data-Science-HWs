**TASK-4B: Please provide the visual illustrations of the dataset that will demonstrate the insightful distributions with respect to the target column (CDRGLOB: the level of the Alzheimer Disease -AD)**

```python
x = data['LATVENT']
y = data['CERECSF']
labels=data['CDRGLOB']
plt.scatter(x,y, cmap='viridis',c=labels, alpha=0.5)
plt.xlabel('Segmented total lateral ventricle volume')
plt.ylabel('Total cerebrum cerebrospinal fluid volume')
plt.title('LATVENT - CERECSF')
plt.colorbar()
plt.show()
```

```python
data.corr()["GRAYVOL"].sort_values(ascending=False).head(3)
```

```python
x = data['GRAYVOL']
y = data['CEREGR']
labels=data['CDRGLOB']
plt.scatter(x,y, cmap='viridis',c=labels, alpha=0.5)
plt.xlabel('Total brain gray matter volume')
plt.ylabel('Total cerebrum gray matter volume')
plt.title('GRAYVOL - CEREGR')
plt.colorbar()
plt.show()
```

```python
x = data ["INDEPEND"]
y = data["HIPPOVOL"]
labels = data["CDRGLOB"]
plt.scatter(x, y, s=10, cmap = "viridis", c=labels, alpha=0.5)
plt.xlabel("Independence")
plt.ylabel("Hippocampus Volume")
plt.title("Independence - HippoVol")
plt.colorbar()
plt.show()
```
